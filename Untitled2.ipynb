첫번째  Mission: 기본 플롯 만들기

```python
import matplotlib.pyplot as plt

def setup_basic_plot(step_num=1, action="forward"):
    fig, ax = plt.subplots(figsize=(10, 8))

    # 축 범위 설정
    ax.set_xlim(-1, 8)
    ax.set_ylim(-1, 12)

    # 라벨 설정
    ax.set_xlabel("X-axis", fontsize=12)
    ax.set_ylabel("Y-axis", fontsize=12)

    # 제목 설정
    ax.set_title(f"Step {step_num}: Action = {action}", fontsize=12)

    return ax

if __name__ == "__main__":
    ax = setup_basic_plot(5, "turn_left")
    plt.tight_layout()
    plt.show()
```

두번째 Mission: 목표선 그리기 함수 만들기

```python
import matplotlib.pyplot as plt

def draw_goal_line(obstacles):
    # 캔버스 설정
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.set_xlim(-1, 8)
    ax.set_ylim(-1, 12)
    ax.grid(True, alpha=0.3)

    # 목표선 계산
    if obstacles:
        max_y = max(y for _, y in obstacles)
    else:
        max_y = 0
    goal_y = max_y + 3

    # 목표선 그리기
    ax.axhline(y=goal_y, color='lime', linewidth=3, alpha=0.7, label=f'GOAL LINE y={goal_y}')

    # 텍스트 표시
    ax.text(7, goal_y + 0.2, f"GOAL (y={goal_y})", color='lime', fontsize=12,
            ha='right', va='bottom')

    # 범례 및 출력
    ax.legend()
    plt.tight_layout()
    plt.show()

    return goal_y

# 실행
obstacles = {(3, 2), (3, 3), (2, 4), (4, 5)}
goal_y = draw_goal_line(obstacles)
print("Returned goal_y:", goal_y)
```

세번째  Mission: 장애물 그리기 함수 만들기

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

def draw_obstacles(obstacles):
    # 캔버스 설정
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.set_xlim(-1, 8)
    ax.set_ylim(-1, 12)
    ax.set_aspect('equal')
    ax.grid(True, alpha=0.3)
    ax.set_xlabel("X-axis", fontsize=12)
    ax.set_ylabel("Y-axis", fontsize=12)
    ax.set_title("Obstacle Map", fontsize=12)

    # 장애물 타입 검사 및 변환
    if isinstance(obstacles, set):
        obstacles = list(obstacles)

    # 장애물 그리기
    for i, (x, y) in enumerate(obstacles):
        # 사각형 추가
        rect = Rectangle((x - 0.4, y - 0.4), 0.8, 0.8, color='red', alpha=0.7,
                         label='Obstacle' if i == 0 else None)
        ax.add_patch(rect)

        # 텍스트(번호) 추가
        ax.text(x, y, str(i+1), color='white', weight='bold',
                ha='center', va='center', fontsize=10)

        # 콘솔 출력
        print(f"Obstacle {i+1}: ({x}, {y})")

    # 범례 및 출력
    ax.legend()
    plt.tight_layout()
    plt.show()

# 실행
obstacles = {(3, 2), (3, 3), (2, 4), (4, 5)}
draw_obstacles(obstacles)
```
